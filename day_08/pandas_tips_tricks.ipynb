# 16- splitting a string into multiple columns

```python
df = pd.DataFrame({
    "name": ["muhma ali", "sheran gul", "Ali Abbas"],
    "location": ["los angles, ca", "Washington, Dc", "Usa"]
})
df
```

```python
df.name.str.split(" ", expand=True)
```

```python
df[["city", "state"]] = df.location.str.split(", ", expand=True)
df
```

```python
df[["first_name", "last_name"]] = df.name.str.split(" ", expand=True)
df
```

```python
df = df[["first_name", "last_name", "city", "state"]]
df
```

Pivoting wide to long format\
***Pandas.melt()*** merges multiple column into one, producing dataframe that is longer than input

```python
df = pd.DataFrame({"key": ["foo", "bar", "baz"],
                    "A": [1, 2, 3],
                    "B": [4, 5, 6],
                    "C": [7, 8, 9]})
```

```python
fasla = df.melt(id_vars='zip', var_name='location', value_name='distance')
fasla
```

```python
# benificial in plotting
```

```python
sns.barplot(data=fasla, x='zip', y='distance', hue='location')
```

```python
sns.barplot(data=df, x='zip', y='factory')
```
